importing panads



```python
import pandas as pd
```

displaying dataset from file

```python
df = pd.read_csv('/content/StudentDeatilsForm.csv')
df
```

for droping single attributes

```python
df = df.drop('Timestamp',axis=1)
```

```python
df
```

for droping multiple aattributes

```python
df = df.drop(['Username','Mobile No'],axis=1)
df
```

for knowing Null values

```python
df.isna().sum()
```

```python
df.info()
```

for converting objests into integer

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['Gender']=le.fit_transform(df['Gender'])
df['Branch']=le.fit_transform(df['Branch'])
df['Result']=le.fit_transform(df['Result'])
df
```

Slicing independent variable


```python
x = df.iloc[:,0:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

Training and testing 
dependent and independent 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

choosing algorithm for classification

```python
from sklearn.ensemble import RandomForestClassifier
```

training data set

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

After training we test using predict and test data

```python
y_pred=rfc.predict(x_test)
```

```python
y_pred
```

Now we make confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.heatmap(cm,
            annot=True,
            xticklabels=['pass','fail'],
            yticklabels=['pass','fail'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

finding accuracy

```python
from sklearn.metrics import accuracy_score 
```

```python
acc1=accuracy_score(y_test,y_pred)
```

```python
acc1
```
